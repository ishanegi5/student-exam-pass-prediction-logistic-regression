```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dropped_dataset=dataset.drop('Student_ID',axis=1)
```

```python
dropped_dataset
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
col_to_encode=['Gender','Internet_Access_at_Home','Extracurricular_Activities','Pass_Fail']
for col in col_to_encode:
  dropped_dataset[col]=le.fit_transform(dropped_dataset[col])
dropped_dataset
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaling_col=['Study_Hours_per_Week','Attendance_Rate','Past_Exam_Scores','Final_Exam_Score']
for col in scaling_col:
    dropped_dataset[col]=sc.fit_transform(dropped_dataset[col].values.reshape(-1,1))
dropped_dataset
```

```python
dropped_dataset.info()
```

```python
final_dataset=dropped_dataset.drop(['Parental_Education_Level'],axis=1)
```

```python
final_dataset
```

```python
x=final_dataset.drop(['Pass_Fail'],axis=1)
y=final_dataset['Pass_Fail']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000,solver='lbfgs')
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accs=accuracy_score(y_test,y_pred)
cfm=confusion_matrix(y_test,y_pred)
clr=classification_report(y_test,y_pred)
```
